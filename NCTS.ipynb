```python
from fenics import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1D simulation
l = 1.0          
A = -0.1         
T0 = 1.0         
Tw = T0 * (1+A)  
nx = 200         
# nt = 1000         
t_total = 1.2   
nt = int(2500 * t_total)

mesh = IntervalMesh(nx, 0, l)
V = FunctionSpace(mesh, 'CG', 1)

T_n = interpolate(Constant(T0), V)   
T_prev = T_n.copy()                  

dt = t_total/nt

T = TrialFunction(V)
v = TestFunction(V)

a = (1/pow(dt,2))*T*v*dx + (1/(2*dt))*T*v*dx + inner(grad(T), grad(v))*dx
L = (2/pow(dt,2))*T_n*v*dx + (1/(2*dt))*T_prev*v*dx - (1/pow(dt,2))*T_prev*v*dx

def boundary(x):
    return near(x[0], 0) or near(x[0], l)
    # return near(x[0], 0)
bc = DirichletBC(V, Constant(Tw), boundary)

T = Function(V)
for n in range(nt):
    solve(a == L, T, bc)
    T_prev.assign(T_n)
    
    T_n.assign(T)
    if n % 20 == 0:
        plot(T)

```

```python
# 2D simulation

from fenics import *
import numpy as np
import matplotlib.pyplot as plt

l_x = 1.0          # Domain length in x
l_y = 1.0          # Domain length in y
A = 5           # Perturbation strength
T0 = 1.0           # Initial temperature
Tw = T0 * (1 + A)  # Boundary temperature
Tw2 = T0*(1 + A)
nx = 200           # Mesh resolution in x
ny = 200           # Mesh resolution in y
t_total = 2      # Total simulation time
nt = int(400 * t_total)
dt = t_total / nt  # Time step
e = np.e

# xdmf_T = XDMFFile("temperature6.xdmf")
# T_data = TimeSeries("temperature6")
# xdmf_q = XDMFFile("heat_flux6.xdmf")
# Q_data = TimeSeries("heatflux6")
# xdmf_T.parameters["flush_output"] = True
# xdmf_q.parameters["flush_output"] = True
# xdmf_T.parameters["functions_share_mesh"] = True
# xdmf_q.parameters["functions_share_mesh"] = True



mesh = RectangleMesh(Point(0, -0.5*l_y), Point(l_x, 0.5*l_y), nx, ny)
V = FunctionSpace(mesh, 'CG', 1)   
W = VectorFunctionSpace(mesh, 'CG', 1)  

T_n = interpolate(Constant(T0), V)  
T_prev = Function(V)                
T_prev.assign(T_n)
q_n = Function(W)                   

T = TrialFunction(V)
v = TestFunction(V)
a_T = (1/pow(dt,2))*T*v*dx + (1/(2*dt))*T*v*dx + inner(grad(T), grad(v))*dx
L_T = (2/pow(dt,2))*T_n*v*dx + (1/(2*dt))*T_prev*v*dx - (1/pow(dt,2))*T_prev*v*dx

q = TrialFunction(W)
w = TestFunction(W)

a_q = inner(q, w)*dx + dt*inner(q, w)*dx
L_q = inner(q_n, w)*dx - dt*inner(grad(T_n), w)*dx

def boundary1(x, on_boundary):
    return (on_boundary and 
            x[1] <= 0.1 and x[1] >= -0.1 and (near(x[0], 0,eps=1e-8)))

def boundary2(x, on_boundary):
    return (on_boundary and 
            x[1] <= 0.1 and x[1] >= -0.1 and near(x[0], l_x,eps=1e-8))

inflow_profile = ('A * cos(pi * x[1] / inflow_height)+B')
inflow_f = Expression(inflow_profile, A=Tw-T0,B=T0, inflow_height=0.2,degree=2)
# inflow_profile = ('T0 + A * pow(B,(-pow((t)/sigma, 2)))')
# inflow_f = Expression(inflow_profile, T0=T0,A=Tw-T0,B=e, t=0,sigma=0.1,degree=2)
bc1 = DirichletBC(V, inflow_f, boundary1)
bc2 = DirichletBC(V, inflow_f, boundary2)
bc=[bc1,bc2]

def update_bc(new_t):
    inflow_f.t = new_t
    bc1 = DirichletBC(V, inflow_f, boundary1)
    bc2 = DirichletBC(V, inflow_f, boundary2)
    bc=[bc1,bc2]
    return bc

T = Function(V)
q = Function(W)

for n in range(nt):
    t = dt * nt
    # bc = update_bc(t)
    solve(a_T == L_T, T, bc)
    
    T_prev.assign(T_n)
    T_n.assign(T)
    
    solve(a_q == L_q, q)
    q_n.assign(q)

    # if n % 20 == 0:
    #     current_time = n * dt
    #     xdmf_T.write(T, current_time)
    #     xdmf_q.write(q, current_time)

    # if n % 10 == 0:
    #     xdmf_T.write(T,t)
    #     T_data.store(T.vector(),t)
    #     xdmf_q.write(q,t)
    #     Q_data.store(q.vector(),t)
    
    
    if n % 10 == 0:
        
        magnitude = project(sqrt(dot(q, q)), V)
        
        coordinates = mesh.coordinates()
        x = coordinates[:, 0]
        y = coordinates[:, 1]
        z = magnitude.compute_vertex_values(mesh)
        
        x_unique = np.sort(np.unique(x))
        y_unique = np.sort(np.unique(y))
        X, Y = np.meshgrid(x_unique, y_unique)
        Z = z.reshape(len(y_unique), len(x_unique))
        
        plt.figure(figsize=(10, 5))
        
        plt.subplot(1, 2, 1)
        t_vals = T.compute_vertex_values(mesh)
        t_plot = t_vals.reshape(len(y_unique), len(x_unique))
        plt.pcolormesh(X, Y, t_plot, cmap='inferno_r', shading='auto')
        plt.colorbar(label='Temperature')
        plt.title(f'Temperature at t = {n*dt:.4f}')
        
        plt.subplot(1, 2, 2)
        plt.pcolormesh(X, Y, Z, cmap='viridis_r', shading='auto')
        plt.colorbar(label='Heat Flux Magnitude')
        plt.title(f'Heat Flux at t = {n*dt:.4f}')
        
        plt.tight_layout()
        plt.show()
        plt.close()


plt.figure(figsize=(10, 6))
plt.pcolormesh(X, Y, Z, cmap='viridis', shading='auto')
plt.colorbar(label='Heat Flux Magnitude')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Final Heat Flux Field (CV Model)')
plt.show()
```
